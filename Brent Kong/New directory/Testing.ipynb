```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

avocadoes = pd.read_csv("./avocado_data/avocado.csv")
```

```python
lin = linear_model.LinearRegression()
lin.fit(x_train, y_train)
y_hat = lin.predict(x_test)
print(y_hat)
#print(lin.intercept_, lin.coef_[0])
```

```python
x2 = avocadoes["Small Bags"].to_numpy()
z2 = avocadoes["Large Bags"].to_numpy()
y2 = avocadoes["Total Volume"].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

x2_train, x2_test, z2_train, z2_test, y2_train, y2_test = train_test_split(x2, z2, y2, test_size=0.2, random_state=0) #none makes the data change every time
```

```python
x2_train = x2_train.reshape(-1,1)
x2_test = x2_test.reshape(-1,1)

y2_train = y2_train.reshape(-1,1)
y2_test = y2_test.reshape(-1,1)

z2_train = z2_train.reshape(-1,1)
z2_test = z2_test.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression() #calls linear regression function
linear.fit(x2_train, y2_train)
print(linear.intercept_, linear.coef_[0])
```

```python
y2_predict = linear.predict(x2_test)

plt.figure(figsize = (10,5))
plt.plot(x2_test, y2_predict ,'-')


plt.scatter(x2_test, y2_test, c = "Red")
plt.title("Number of Small Bags vs. Total Volume of Avocadoes", fontsize = "18")
plt.xlabel("Number of Small bags",fontsize = "15")
plt.ylabel("Total Avocado Volume", fontsize = "15")
plt.grid("on")
plt.show
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression() #calls linear regression function
linear.fit(z2_train, y2_train)
print(linear.intercept_, linear.coef_[0])
```

```python
z2_predict = linear.predict(z2_test)

plt.figure(figsize = (10,5))
plt.plot(z2_test, z2_predict ,'-')


plt.scatter(z2_test, y2_test, c = "Pink")
plt.title("Number of Large Bags vs. Total Volume of Avocadoes", fontsize = "18")
plt.xlabel("Number of Large bags",fontsize = "15")
plt.ylabel("Total Avocado Volume", fontsize = "15")
plt.grid("on")
plt.show
```
